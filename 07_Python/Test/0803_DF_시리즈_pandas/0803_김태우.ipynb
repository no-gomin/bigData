```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
```

```python
# 임의의 행에 인위적으로 결측치 할당
import random
random.seed(12)
for col in range(iris.shape[1]-1) : 
    iris.iloc[random.sample(range(len(iris)),20),col] = np.nan
```

```python
iris.isnull().sum()
```

```python
# 결측치를 평균값으로 대체 (열마다 평균값 -> 결측치를 평균값으로 대체)
iris = iris.fillna(value=iris.mean())
```

```python
# 결측치가 평균값으로 대체되었는지 확인
iris.isnull().sum()
```

```python
# 1. 처음 다섯개행
iris.head(5)
```

```python
# 2. 독립변수 x, 종속변수 y로 나누기
X = iris.iloc[:,:-1]
X
```

```python
# 3. 처음 50개 행을 temp 변수에
temp = iris[:50]
temp
```

```python
# 5. versicolor종의 데이터만 iris_versicolor 변수에
iris_versicolor = iris.loc[iris['species']=='versicolor']
iris_versicolor
```

```python
# 6. 2번의 X와 Y를 합하여 iris_df 데이터프레임 생성
iris_df = pd.concat([X,Y], axis=1)
iris_df.info()
```

```python
# 7. iris 데이터의 각 열의 평균
iris.iloc[:,0:4].apply(np.mean, axis=0)
```

```python
# 8. iris 데이터의 각 열들 사이의 상관계수
r = iris.corr()
r
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(r, annot=True, cmap='Blues', annot_kws={'size':15})
plt.show()
```

```python
# 9. iris 데이터의 각 요소들과 변수별 평균과의 차이 + 다섯개행
temp_9_X = iris.iloc[:,:-1]
temp_9_mean = temp_9_X.apply(np.mean)
temp_9_result = temp_9_X.apply(lambda x : x-temp_9_mean, axis=1)
temp_9_result.head(5)
```
