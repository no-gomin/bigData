# 2. daum의 많이 본 뉴스 랭킹 수집 및 저장

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://news.daum.net/ranking/popular'
news_best = requests.get(url)
soup = BeautifulSoup(news_best.content, 'html.parser')
```

```python
news_title = soup.select('ul.list_news2 > li > div.cont_thumb > strong.tit_thumb > a.link_txt')
news_comname = soup.select('div.cont_thumb > strong.tit_thumb > span.info_news')
len(news_title), len(news_comname)
```

```python
best_list = []
print('newstitle, newscomname', end = '\n\n')
for idx in range(len(news_title)) :
    print('{}, {}'.format(news_title[idx].text, news_comname[idx].text))
    dic = {}
    dic['newstitle'] = news_title[idx].text
    dic['newscomname'] = news_comname[idx].text
    best_list.append(dic)
df = pd.DataFrame(best_list)
df.to_csv('news.csv')
```
